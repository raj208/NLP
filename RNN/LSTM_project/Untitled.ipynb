```python
from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.preprocessing.text import one_hot
```

```python
voc_size = 10000
msg = X.copy()
```

```python
msg['review'][2]
```

```python
import nltk
import re
from nltk.corpus import stopwords
```

```python
nltk.download('stopwords')
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
corpus=[]
for i in range(0,len(msg)):
    review = re.sub('[^a-zA-Z]', ' ', msg['review'][i])
    review=review.lower()
    review = review.split()
    
    review = (ps.stem(word) for word in review if word in stopwords.words('english'))
    review = ' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
onehot_repr = [one_hot(words, voc_size) for words in corpus]
```

```python
onehot_repr
```

```python
max_length = max(len(sentence.split()) for sentence in corpus)

print(f"The maximum length of the corpus is: {max_length}")
```

```python
msg_len = 1210
embedded_docs = pad_sequences(onehot_repr, padding='pre', maxlen = msg_len)
```

embedded_docs

```python
embedding_dim = 300
model=Sequential()
model.add(Embedding(voc_size, embedding_dim, input_length= msg_len))
model.add(LSTM(200))
model.add(Dense(1, activation = "sigmoid"))
model.compile(loss='binary_crossentropy', optimizer = 'adam', metrics=['accuracy'])
```

```python
model.summary()
```

```python
X_final = np.array(embedded_docs)
y_final = np.array(y)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_final, y_final, test_size = 0.3, random_state = 42)
```

```python
print(type(X_train), X_train.shape)
print(type(y_train), y_train.shape)
```

```python
model.fit(X_train, y_train, validation_data = (X_test, y_test), epochs = 20, batch_size = 16)
```
